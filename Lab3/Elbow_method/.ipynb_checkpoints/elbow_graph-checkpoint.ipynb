```python
from matplotlib import pyplot as plt
import pandas as pd
from IPython.display import set_matplotlib_formats
%matplotlib inline
set_matplotlib_formats('svg')
legend_properties = {'weight':'bold'}

data = pd.read_csv('test.txt')

print(data)
plt.plot(data['k'], data['variance'],linewidth=2)
# plt.scatter([5], [146.348093], color='b')
plt.scatter( 4 , 1.320579  , s=1000 ,  facecolors='none', edgecolors='green',linewidth=2 )
plt.xticks(fontweight='bold')
plt.yticks(fontweight='bold')
plt.xlabel('K',fontweight='bold',fontsize=12)
plt.ylabel("Variance",fontweight='bold',fontsize=12)
plt.title("2D Points Dataset",fontweight='bold')
plt.show
# print(data)
```
