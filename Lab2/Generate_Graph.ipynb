```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
from operator import truediv
from IPython.display import set_matplotlib_formats
%matplotlib inline
set_matplotlib_formats('svg')

legend_properties = {'weight':'bold'}

Data = pd.read_csv('StatisticalData/car',header=0)
Data.head(2)
group = Data.groupby('Classifier')
NaiveBayes = group.get_group('NaiveBayes')
DecisionTree = group.get_group('DecisionTree')

Avg_Naive = NaiveBayes.groupby('testSize').mean()
Avg_Decision = DecisionTree.groupby('testSize').mean()
print(NaiveBayes.head(3))
print(DecisionTree.head(3))
```

```python
x_axis = [0.1,0.2,0.3,0.4]

resultType = 'accuracy'

plt.plot(x_axis,Avg_Naive[resultType],'-o',label = 'Naive Bayes',linewidth=2)
plt.plot(x_axis,Avg_Decision[resultType],'-o',label = 'Decision Tree',linewidth=2)

plt.xlim(0.42,0.08)
plt.xticks(x_axis,fontweight='bold')
plt.yticks(fontweight='bold')
plt.xlabel('Test Size',fontweight='bold',fontsize=12)
plt.ylabel(resultType.title()+" %",fontweight='bold',fontsize=12)
plt.text(.4, 87.5, 'Dataser: Car', style='normal',fontweight='bold')
plt.legend(prop=legend_properties) 

plt.savefig("Car_"+str(resultType))
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd 
from operator import truediv
from IPython.display import set_matplotlib_formats
%matplotlib inline
set_matplotlib_formats('svg')

legend_properties = {'weight':'bold'}

Data = pd.read_csv('StatisticalData/bank',header=0)
Data.head(2)
group = Data.groupby('Classifier')
NaiveBayes = group.get_group('NaiveBayes')
DecisionTree = group.get_group('DecisionTree')

Avg_Naive = NaiveBayes.groupby('testSize').mean()
Avg_Decision = DecisionTree.groupby('testSize').mean()

print("Decision")
print(Avg_Decision.head(3))

print("Naive")
print(Avg_Naive.head(3))
```

```python
n_groups = 4
means_frank = (92.666667,  92.418803,  92.540497,  92.206628)
means_guido = (97.333333,  96.806878,  97.034799,  96.853708)

# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.35
opacity = 0.8

rects1 = plt.bar(index, means_frank, bar_width,
alpha=opacity,
label='Decision Tree')

rects2 = plt.bar(index + bar_width, means_guido, bar_width,
alpha=opacity,
label='Naive Bayes')

plt.xlabel('Measure Name',fontweight='bold',fontsize=12)
plt.ylabel('Scores',fontweight='bold',fontsize=12)
plt.ylim(0,119)
plt.title('Dataset: Iris',fontweight='bold',fontsize=12)
plt.xticks(index + bar_width, ('Accuracy', 'Precision', 'Recall', 'F measure'))
plt.legend()

plt.tight_layout()
plt.show()
```
